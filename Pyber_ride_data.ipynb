```python
# Import Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statistics as st
```

```python
# Read CSV file into dataframe
df = pd.read_csv("Resources/PyBer_ride_data.csv")
df
```

```python
# Set x-axis and tick locations.
# First, we need an array that is the length of the DataFrame. This will allow us to plot each month for each value 
# in the array. With the NumPy module, we create an array of numbers using the arange function for the df
x_axis = np.arange(len(df))

# Then we set the tick locations for every value, or month, in the x-axis using list comprehension.
tick_locations = [value for value in x_axis]

# Label x & y axes
# We add the xticks() function and pass in the tick_locations and the labels of the x-ticks, df["Month"].
df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, df["Month"])
```
