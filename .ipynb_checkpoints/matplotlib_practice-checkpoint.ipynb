```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total far in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot (using MATLAB)
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() (Using OOI: Method #1)
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() (Using OOI: Method #2)
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Creat the plot with ax.plt() (Using OOI: Method #3) To change figure-level attributes(such as axis labels, a title, or a 
# legend) or save the figure as an image, use fig = plt.figure()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot w/ plt.show()
plt.plot(x_axis, y_axis)
plt.show()
```

```python
plt.plot(x_axis, y_axis)
```

```python
# Create the plot and add a lobel for the legend.
plt.plot(x_axis, y_axis, label="Boston")

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit to between 0 and 45
plt.ylim(0, 45)

# Create a title
plt.title("PyBer Fare by Month")

# Add a legend
plt.legend()
```

```python
# Create the plot and amend the code bloc to add a new line for color, modified line width, and a marker for our data points.
# Using MATLAB methods
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label="Boston")

# Create labels for x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")

#Set the y limit between 0 and 45
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid
plt.grid()

# Add the legend.
plt.legend()
```

```python
# Create the plot and amend the code bloc to add a new line for color, modified line width, and a marker for our data points.
# Using OOI methods
ax = plt.axes()
ax.plot(x_axis, y_axis, marker="*", color="green", linewidth=2, label="Boston")

# Create labels for x and y axes.
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")

#Set the y limit between 0 and 45
ax.set_ylim(0, 45)

# Create a title.
ax.set_title("PyBer Fare by Month")

# Add a grid
ax.grid()

# Add the legend.
ax.legend()
```

```python
# Create bar chart using ML
plt.bar(x_axis, y_axis)
```

```python
# Create the bar chart using ML with annotation
plt.bar(x_axis, y_axis, color="green", label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
# Create a horizontal bar chart with inverted axes
plt.barh(y_axis, x_axis)
```

```python
# Create a horizontal bar chart like OUT[54] but with the months on y-axis inverted
# Using gca() to invert.
    # gca() method means "get current axes"
    # chaining gca() to invert_yaxis()
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
# Create a horizontal bar chart
plt.barh(x_axis, y_axis, color="cyan", label="Chicago")
plt.gca().invert_yaxis()

#Add a legend
plt.legend()

# Create labels for the x and y axes.
plt.ylabel("Date")
plt.xlabel("Fare($)")

# Create a title.
plt.title("PyBer Fare by Month")
```

```python
# Plot scatterplot using MATLAB (plt.plot())
plt.plot(x_axis, y_axis, "o")
```

```python
# Plot scatterplot using MATLAB (plt.scatter())
plt.scatter(y_axis, x_axis)

```

```python
# Make a scatter plot using MATLAB, plot approach
plt.plot(x_axis, y_axis, c="r", label="Chicago")

# Invert y-axis
plt.gca().invert_yaxis

# Add a legend
plt.legend()

# Add labels for x and y axes
plt.ylabel("Month")
plt.xlabel("Fare")

# Add title
plt.title("PyBer Fare by Month")

```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)

# Create a bubble chart using MATLAB method
plt.scatter(x_axis, y_axis, s=y_axis_larger)

```

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Create bubble chart using OOI method
# chaning color to sky blue ("xkcd:sky blue")
# changing size of bubbles by iterating through y-axis to multiply data by 5
# change transparency to 20% (alpha=)
# iterate through y-axis (for i in y_axis)
bigger_y = []
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i*5 for i in y_axis], c="xkcd:sky blue", alpha=0.2, edgecolors="black", linewidth=2, label="Boston")

# Add a legend
plt.legend()

# Invert y-axis
plt.gca().invert_yaxis

# Label x & y axes
plt.xlabel("Fare ($)")
plt.ylabel("Months")

# Add title
plt.title("PyBer Fare by Month")

# Add min/max to x-axis
plt.xlim(0, 50)


```

```python
# Create a pie chart using MATLAB method
plt.pie(y_axis, labels=x_axis)
```

```python
# Create a pie chart using MATLAB method
# Jazz up pie chart with "exploding wedges" tuple
# Spice up the colors with a color tuple and a unique color ass'n to each wedge
# Use autopct to add % to each wedge and provide the format of 1 decimal place using .1f%
# Increase the size of the pie chart to make easier to read using plt.subplots(figsize=(8, 8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, explode = explode_values, colors=colors, labels=x_axis, autopct='%.1f%%')

```

```python
# Create a pie chart using OOI
fig, ax = plt.subplots()
ax.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Create a pie chart using OOI and adding edits ass'n in 5.1.8
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
fig, ax = plt.subplots(figsize = (8, 8))
ax.pie(y_axis, labels=x_axis, explode = explode_values, colors = colors, counterclock = False, autopct='%.1f%%', shadow=True, startangle = 95)
plt.show()

```
